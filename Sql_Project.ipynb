## <font color="navy blue"> 1. üíµ Total Cost Estimation
### <font color="191970"> Calculate total estimated cost of education for students

```python
pysqldf = lambda q: sqldf(q, globals())

total_cost = """
SELECT DISTINCT Country, University,Level, Program, Duration_Years, Tuition_USD,
       Total_Rent_Cost, Visa_Fee_USD, Insurance_USD, Total_Cost_USD
FROM (
    SELECT
      Country,
      Level,
      University,
      Program,
      Duration_Years,
      Tuition_USD,
      (Rent_USD * Duration_Years * 12) AS Total_Rent_Cost,
      Visa_Fee_USD,
      Insurance_USD,
      (Tuition_USD + (Rent_USD * Duration_Years * 12) + Visa_Fee_USD + Insurance_USD) AS Total_Cost_USD
    FROM
      df
)

ORDER BY
  Country asc;

  """
result = pysqldf(total_cost)
result
```

üîç Findings:
1. Countries show wide variation in total study cost. Countries like Germany or France tend to have lower tuition,    while USA or UK have higher total costs due to both tuition and living expenses.
2. Longer programs (e.g., Bachelor's) significantly increase rent and insurance costs, even if tuition is low.
3. Postgraduate programs often have shorter durations but higher tuition. Undergraduate programs have higher          living expenses due to extended duration.

##  <font color="navy blue">üí≤ Cheapest Countries for a Program
### <font color="191970"> Find the most affordable country for "Data Analytics" program (based on total cost)

```python
pysqldf = lambda q: sqldf(q, globals())

cheapest_program = """
SELECT
  Country,
  University,
  Level,
  Program,
  Duration_Years,
  Tuition_USD,
  (Rent_USD * Duration_Years * 12) AS Rent_Cost,
  (Tuition_USD + (Rent_USD * Duration_Years * 12) + Visa_Fee_USD + Insurance_USD) AS Total_Cost
FROM
  df
WHERE
  Program = 'Data Analytics'
ORDER BY
  Total_Cost ASC
LIMIT 5;

   """
result = pysqldf(cheapest_program)
result
```

üîç Findings:
1. The query successfully highlights the 5 least expensive Data Analytics programs, providing a clear list for        budget-conscious students.
2. The result helps identify countries/universities that offer value-for-money education in a high-demand field       like Data Analytics.
3. These universities are typically located in countries with:

   Lower tuition fees   ,    Affordable rent   and    Lower visa and insurance charges

## <font color="navy blue"> 8. üí∏Total Estimated Monthly Expense
### <font color="191970">  Find monthly expenditure

```python
pysqldf = lambda q: sqldf(q, globals())
monthly_expense = """

SELECT
  u.University,
  u.Country,
  u.City,
  p.Program,
  ROUND((co.Rent_USD + (co.Insurance_USD / (p.Duration_Years * 12))), 2) AS Estimated_Monthly_Expense
FROM
  costs co
JOIN universities u ON co.University_ID = u.University_ID
JOIN programs p ON co.Program_ID = p.Program_ID
ORDER BY
  Estimated_Monthly_Expense ASC
LIMIT 10;
"""
pysqldf(monthly_expense)
```

üîç Findings:

1. Students seeking budget-friendly international education can strongly consider universities in Tunisia, Egypt,     Bangladesh, and Algeria, where monthly living expenses remain under $200 USD‚Äîan attractive option for both         local and global students.

2. These affordable universities offer diverse programs, including Artificial Intelligence, Computer Science,         Information Systems, and Data Analytics, showing that cost-effectiveness does not compromise program choice.

3. Countries like Tunisia, Egypt, and Bangladesh consistently appear in the top 10, reflecting affordable housing     and insurance rates.

## <font color="navy blue">  9. üìà Top 5 Most Expensive Programs by Total Cost
### <font color="191970"> Find Countries where the cost of living and studying is Expensive

```python
pysqldf = lambda q: sqldf(q, globals())
expensive = """

SELECT
  u.University,
  u.Country,
  p.Program,
  Tuition_USD,
  (co.Tuition_USD + co.Rent_USD * p.Duration_Years * 12 + co.Visa_Fee_USD + co.Insurance_USD) AS Total_Cost
FROM
  programs p
JOIN costs co ON p.Program_ID = co.Program_ID
JOIN universities u ON u.University_ID = co.University_ID
ORDER BY
  Total_Cost DESC
LIMIT 5;
"""
pysqldf(expensive)
```

üîç Findings:

1. Stanford University (USA) dominates the list of most expensive programs, with Computer Science and Physics         programs incurring total costs over $190,000 USD.

2. Top-tier U.S. universities not only charge premium tuition fees but also have significant additional living and    administrative costs, making them some of the costliest education destinations globally.

3. These high costs are due to a combination of high tuition fees, long program durations, and substantial            living/visa/insurance expenses.

## <font color="navy blue"> 11. üìã List of Programs Available in More Than 5 Countries
### <font color="191970">  Global popularity of specific programs.

```python
pysqldf = lambda q: sqldf(q, globals())
program = """
SELECT
  p.Program,
  COUNT(DISTINCT u.Country) AS Countries_Offered
FROM
  programs p
JOIN costs co ON p.Program_ID = co.Program_ID
JOIN universities u ON co.University_ID = u.University_ID
GROUP BY
  p.Program
HAVING
  COUNT(DISTINCT u.Country) > 5
ORDER BY
  Countries_Offered DESC;
  """
pysqldf(program)
```

üîç Findings:

1. Computer Science is the most widespread program, offered in 56 countries, indicating its universal demand and      relevance.

2. Tech-focused programs dominate the list, reflecting global trends in digitalization, automation, and data-         driven industries.

3. The results underscore that Computer Science and its related disciplines are not only popular but also widely      accessible globally‚Äîmaking them strategic choices for students aiming for global academic and career mobility.

## <font color="navy blue"> 12. üí±Currency Exchange Impact
### <font color="191970"> Adjust tuition based on exchange rate for foreign students paying in local currency.

```python
pysqldf = lambda q: sqldf(q, globals())
local_currency = """
SELECT
  Country,
  University,
  Tuition_USD,
  Exchange_Rate,
  ROUND(Tuition_USD * Exchange_Rate, 2) AS Local_Currency
FROM
  costs co
  join universities u ON u.University_ID = co.University_ID;
  """
pysqldf(local_currency)
```

üîç Findings:

1. Countries with stronger currencies (e.g., UK, Eurozone) have tuition fees that appear lower when converted to      local currency.

2. In contrast, countries with weaker currencies (e.g., Malaysia, Saudi Arabia) see a much higher local currency      value for the same USD fee

3. This analysis is critical for international students evaluating the real cost of education abroad. It              highlights that while tuition in USD might seem comparable, the financial burden in local currency varies          drastically based on the exchange rate‚Äîmaking some countries more economically accessible than others.

 ## <font color="navy blue"> 13. üè° Find Countries Where Rent Is Over 100000 in Local Currency

```python
pysqldf = lambda q: sqldf(q, globals())
query = """
SELECT
  u.Country,
  u.University,
  co.Rent_USD,
  (co.Rent_USD * co.Exchange_Rate) AS Rent_Local
FROM
  costs co
JOIN universities u ON co.University_ID = u.University_ID
WHERE
  (co.Rent_USD * co.Exchange_Rate) > 100000
ORDER BY
  Rent_Local DESC
  """
pysqldf(query)
```

üîç Findings:

1. All top entries are Iranian universities (e.g., Sharif University of Technology, Isfahan University of Technology).

2. Due to a very high exchange rate (1 USD ‚âà 42,150 IRR), even modest rent in USD (~$300‚Äì400) translates into millions of Iranian Rials, with the top entry exceeding 16.8 million IRR.

3. While rent in USD may seem affordable, currency conversion can significantly alter its impact. Countries like Iran exhibit skyrocketing local rent costs, not due to actual rent inflation, but because of currency devaluation. This highlights the importance of analyzing costs in both USD and local currency when evaluating affordability.

4. This analysis shows how high exchange rates drastically inflate perceived costs in the local context, which is critical for budgeting and financial planning for international students.

## <font color="navy blue"> 14. üíπHIGHEST EXCHANGE RATE 
### <font color="191970"> What Currency Is Strongest Against the Dollar

```python
pysqldf = lambda q: sqldf(q, globals())
highest_exchange_rate = """
SELECT
  Country,
  Exchange_Rate,
  Tuition_USD,
  (co.Tuition_USD * co.Exchange_Rate) as local
FROM
  costs as co
  JOIN universities u ON co.University_ID = u.University_ID
group by Exchange_Rate
ORDER BY
  Exchange_Rate asc
  limit 10;
"""
pysqldf(highest_exchange_rate)
```

üîç Findings:
1. Denmark (0.15), Kuwait (0.31), and Bahrain (0.38) have the lowest exchange rates, meaning their currencies are stronger than the USD.

2. Because of the strong currency, even higher USD tuition fees result in relatively lower local currency amounts.

3. Countries like Denmark, Kuwait, and Bahrain have strong currencies, so even high tuition in USD converts to lower local currency amounts, which could influence affordability for local students. On the other hand, countries like Australia and Canada show the opposite trend, with tuition inflating when seen in local currency terms. 

## <font color="navy blue"> 15.üìâ                               LOWEST EXCHANGE RATE 
### <font color="191970">  What Currency Is Weakest Against the Dollar

```python
pysqldf = lambda q: sqldf(q, globals())
lowest_exchange_rate = """
SELECT
  Country,
  Exchange_Rate,
  Tuition_USD,
  (co.Tuition_USD * co.Exchange_Rate) as local
FROM
  costs as co
  JOIN universities u ON co.University_ID = u.University_ID
group by Exchange_Rate
ORDER BY
  Exchange_Rate desc
LIMIT 10;
"""
pysqldf(lowest_exchange_rate)
```

üîç Findings:

1. Countries like Iran (42,150), Vietnam (24,450), and Lebanon (15,000) have very high exchange rates.
This means 1 USD = 42,150 Iranian Rials or 15,000 Lebanese Pounds.

2. Though tuition fees in USD are moderate (e.g., $3,000‚Äì$5,000), the equivalent in local currency becomes millions.
Iran: $3,000 ‚Üí 126,450,000 Rials
Lebanon: $20,000 ‚Üí 300,000,000 LBP

3. Countries with the highest exchange rates face very high tuition fees in their local currencies, even when the USD value appears affordable. This reflects the currency devaluation and inflation challenges in many developing nations, which can make global education far more expensive for local students.

## <font color="navy blue"> 16. üèÖRank Universities by Tuition Fee per Country

```python
pysqldf = lambda q: sqldf(q, globals())
Rank = """
SELECT
  u.Country,
  u.University,
  p.Program,
  co.Tuition_USD,
  RANK() OVER (PARTITION BY u.Country ORDER BY co.Tuition_USD DESC) AS Tuition_Rank
FROM
  costs co
JOIN universities u ON co.University_ID = u.University_ID
JOIN programs p ON co.Program_ID = p.Program_ID;
"""
pysqldf(Rank)
```

Findings:
Each country has been assigned a Rank 1 university, representing the highest tuition fee institution for that region.
